```python
data['sex'] = data['sex'].replace('male', 1)
data['sex'] = data['sex'].replace('female', 0)
data
```

```python
data['smoker'] = data['smoker'].replace('yes',1)
data['smoker'] = data['smoker'].replace('no', 0)
data
```

```python
data['is_north'] = [1 if x[:5] == 'north' else 0 for x in data['region']]
data['is_east'] = [1 if x[-4:] == 'east' else 0 for x in data['region']]
data
```

```python
cleandata = data[data.columns.difference(['region'])]
cleandata
```
